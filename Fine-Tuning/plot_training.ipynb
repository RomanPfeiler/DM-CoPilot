```python
# notebooks/plot_training.ipynb or just a cell
import pandas as pd
import matplotlib.pyplot as plt
import os

# --- PATH CONFIGURATION ---
RESULTS_PATH = r"c:\Users\pfeil\My Drive\Studys\CAS-NLP-Uni-Bern\Module 4  NLP Transformers\dungeon_master_copilot\Output\fine_tuned_qwen_dnd\eval\Information-Retrieval_evaluation_dnd_validation_results.csv"

print(f"Looking for file at: {RESULTS_PATH}")

if not os.path.exists(RESULTS_PATH):
    print(f"Error: File not found at {RESULTS_PATH}")
else:
    # --- 2. LOAD DATA ---
    df = pd.read_csv(RESULTS_PATH)
    
    # --- 3. DEFINE EXACT COLUMNS ---
    # Based on the header you provided:
    acc_col = 'cosine-Accuracy@1'
    mrr_col = 'cosine-MRR@10'

    print(f"Plotting columns: '{acc_col}' and '{mrr_col}'")

    # --- 4. PLOT ---
    plt.figure(figsize=(10, 5))
    
    # Plot Accuracy (Solid Line)
    plt.plot(df['epoch'], df[acc_col], label='Accuracy@1 (Perfect Match)', marker='o', linewidth=2)
    
    # Plot MRR (Dashed Line)
    plt.plot(df['epoch'], df[mrr_col], label='MRR@10 (General Rank)', marker='s', linestyle='--', linewidth=2)
    
    plt.xlabel('Epochs')
    plt.ylabel('Score (0.0 - 1.0)')
    plt.title('Fine-Tuning Performance: Qwen-0.6B on Rules')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Force the Y-axis to show 0 to 1 range clearly
    plt.ylim(0, 1.1) 
    
    plt.show()

    print("\nInterpreting the Graph:")
    print(f"Current Accuracy: {df[acc_col].iloc[-1] * 100:.1f}%")
    print("If this is 1.0 (100%), your model perfectly memorized the small test set!")
```
